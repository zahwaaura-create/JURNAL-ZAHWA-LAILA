```python
import io
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['smmh.csv']))
print(f'Berhasil! Total data: {len(df)}')
```

```python
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

# Filter Gen Z 18-27
df = df[(df['1. What is your age?'] >= 18) & (df['1. What is your age?'] <= 27)].copy()
df = df.reset_index(drop=True)
print(f'Setelah filter Gen Z: {len(df)}')

# Buat label
gejala_cols = [
    '13. On a scale of 1 to 5, how much are you bothered by worries?',
    '18. How often do you feel depressed or down?',
    '20. On a scale of 1 to 5, how often do you face issues regarding sleep?',
    '14. Do you find it difficult to concentrate on things?',
    '12. On a scale of 1 to 5, how easily distracted are you?'
]
df['skor_rata'] = df[gejala_cols].mean(axis=1)
```

```python
df['Label'] = df['skor_rata'].apply(lambda x: 'Ya' if x >= 3.5 else 'Tidak')
df = df.drop(columns=['skor_rata', 'Timestamp'])
print(f'Ya: {len(df[df["Label"]=="Ya"])}, Tidak: {len(df[df["Label"]=="Tidak"])}')

# Missing value
df['5. What type of organizations are you affiliated with?'] = df['5. What type of organizations are you affiliated with?'].fillna(
    df['5. What type of organizations are you affiliated with?'].mode()[0])

# Balancing
ya = df[df['Label'] == 'Ya']
tidak = df[df['Label'] == 'Tidak']
ya_sample = ya.sample(n=len(tidak), random_state=42)
df = pd.concat([ya_sample, tidak]).sample(frac=1, random_state=42).reset_index(drop=True)
print(f'Setelah balancing: {len(df)} (Ya:{len(df[df["Label"]=="Ya"])}, Tidak:{len(df[df["Label"]=="Tidak"])})')
```

```python
cat_cols = ['2. Gender','3. Relationship Status','4. Occupation Status',
            '5. What type of organizations are you affiliated with?',
            '6. Do you use social media?',
            '7. What social media platforms do you commonly use?',
            '8. What is the average time you spend on social media every day?']

le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col].astype(str))

# Normalisasi
X = df.drop(columns=['Label'])
y = df['Label']
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Seleksi fitur
selector = SelectKBest(f_classif, k=10)
X_selected = selector.fit_transform(X_scaled, y)

# Split 70:30
X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y, test_size=0.3, random_state=42, stratify=y)
print(f'Data latih: {len(X_train)}, Data uji: {len(X_test)}')
```

```python
# Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)

# Evaluasi
cm = confusion_matrix(y_test, y_pred, labels=['Ya','Tidak'])
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, pos_label='Ya', zero_division=0)
rec = recall_score(y_test, y_pred, pos_label='Ya', zero_division=0)

TP=cm[0][0]; FN=cm[0][1]; FP=cm[1][0]; TN=cm[1][1]

print("\n=== HASIL KLASIFIKASI NAÏVE BAYES ===")
print(f"\nCONFUSION MATRIX:")
print(f"                  Prediksi Ya   Prediksi Tidak")
print(f"Aktual Ya    :        {TP}              {FN}")
print(f"Aktual Tidak :         {FP}              {TN}")
print(f"\nTP={TP} | FN={FN} | FP={FP} | TN={TN}")
print(f"\nAkurasi  : {acc*100:.2f}%")
print(f"Presisi  : {prec*100:.2f}%")
print(f"Recall   : {rec*100:.2f}%")
```
